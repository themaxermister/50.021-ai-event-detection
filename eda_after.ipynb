```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/test/output_with_category.csv')
```

```python
print(f"Number of Unique Categories: {len(df['category'].unique())}")
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['word_count'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Word Count')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of word count for each category show category name

plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='word_count', data=df)
plt.title('Distribution of Word Count for Each Category')
plt.xlabel('Category')
plt.ylabel('Word Count')
plt.xticks(rotation=45)
plt.gca().set_xticklabels(df['category'].unique(), rotation=45, ha='right')
plt.show()
```

```python
# Get the top 10 repeated occurrences of the 'type' column
top_10_types = df['category'].value_counts()

# Visualize top 10 types of news articles
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=top_10_types.index, y=top_10_types.values)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')
plt.title("Types of News Articles", size=15)
plt.xlabel("Types of Articles", size=14)
plt.ylabel("Number of Articles", size=14)
plt.xticks(rotation=45)
plt.show()
```
